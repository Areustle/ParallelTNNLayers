```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
```

```python
def plot3dRank(cp_df, nv_df, Var, x_label):
#     fig = plt.figure(figsize=plt.figaspect(0.5))
    fig = plt.figure(figsize=(16, 9), dpi=128)
    fig.suptitle('2D Convolution - ' + x_label, fontsize=16)
    
    ax_cp = fig.add_subplot(1,2,1, projection='3d')
    ax_cp.plot_trisurf(cp_df[Var], cp_df['fRank'], cp_df['ns'], cmap=plt.cm.viridis, linewidth=0.2)
    ax_cp.set_title('CP Decomposed')
    ax_cp.set_xlabel(x_label)
    ax_cp.set_ylabel('Tensor Rank')
    ax_cp.set_zlabel('Execution Time (ns)')
    
    ax_nv = fig.add_subplot(1,2,2, projection='3d')
    ax_nv.plot_trisurf(nv_df[Var], nv_df['fRank'], nv_df['ns'], cmap=plt.cm.viridis, linewidth=0.2)
    ax_nv.set_title('cuDNN')
    ax_nv.set_xlabel(x_label)
    ax_nv.set_ylabel('Tensor Rank')
    ax_nv.set_zlabel('Execution Time (ns)')
    
    plt.show()
```

```python
nv_df = pd.read_csv("NVConv2dForward_results_2070_batch_size.txt")
cp_df = pd.read_csv("CP4Conv2dForward_results_2070_batch_size.txt")
plot3dRank(cp_df, nv_df, 'N', 'Batch Size')
```

```python
cp_df = pd.read_csv("CP4Conv2dForward_results_2070_image_size.txt")
nv_df = pd.read_csv("NVConv2dForward_results_2070_image_size.txt")
plot3dRank(cp_df, nv_df, 'H', 'Image Height')
```

```python
cp_df = pd.read_csv("CP4Conv2dForward_results_2070_channel_depth.txt")
nv_df = pd.read_csv("NVConv2dForward_results_2070_channel_depth.txt")
plot3dRank(cp_df, nv_df, 'C', 'Channel Depth')
```

```python
cp_df = pd.read_csv("CP4Conv2dForward_results_2070_filter_count.txt")
nv_df = pd.read_csv("NVConv2dForward_results_2070_filter_count.txt")
plot3dRank(cp_df, nv_df, 'fK', 'filter count')
```
